```python
import pandas as pd
import numpy as np
from random import choice, randint, uniform

np.random.seed(42)

brands = ['Apple','Samsung','Dell','HP','Lenovo','Asus','Acer','Sony','LG','Xiaomi','OnePlus','Intel','Cisco','Nokia','Panasonic']
device_types = {
    'Smartphone':['iPhone 11','Galaxy S21','Redmi Note','OnePlus 9'],
    'Laptop':['MacBook Pro','Dell XPS','HP Pavilion','Lenovo ThinkPad'],
    'Desktop':['Gaming PC','Office PC','Workstation'],
    'Tablet':['iPad','Galaxy Tab','Mi Pad'],
    'TV':['Smart TV','LED TV','OLED TV'],
    'Router':['WiFi Router','Enterprise Router'],
    'Server':['Rack Server','Blade Server'],
    'Printer':['Laser Printer','Inkjet Printer']
}
conditions = ['Excellent','Good','Fair','Damaged','Scrap']
cond_factor = {'Excellent':1.0,'Good':0.92,'Fair':0.8,'Damaged':0.62,'Scrap':0.45}
base = {
    'Smartphone': {'gold':0.035,'silver':0.25,'copper':18,'palladium':0.012,'platinum':0.001,'cost':8},
    'Laptop': {'gold':0.09,'silver':0.9,'copper':140,'palladium':0.03,'platinum':0.002,'cost':18},
    'Desktop': {'gold':0.12,'silver':1.4,'copper':260,'palladium':0.04,'platinum':0.003,'cost':24},
    'Tablet': {'gold':0.05,'silver':0.45,'copper':42,'palladium':0.015,'platinum':0.001,'cost':10},
    'TV': {'gold':0.02,'silver':0.3,'copper':95,'palladium':0.006,'platinum':0.001,'cost':16},
    'Router': {'gold':0.015,'silver':0.2,'copper':55,'palladium':0.01,'platinum':0.001,'cost':9},
    'Server': {'gold':0.22,'silver':2.8,'copper':420,'palladium':0.08,'platinum':0.005,'cost':40},
    'Printer': {'gold':0.01,'silver':0.12,'copper':70,'palladium':0.004,'platinum':0.0005,'cost':12}
}
brand_factor = {b: np.random.uniform(0.94,1.08) for b in brands}
rows=[]
for _ in range(10000):
    dtype = choice(list(device_types.keys()))
    item = choice(device_types[dtype])
    brand = choice(brands)
    condition = choice(conditions)
    year = randint(2008,2025)
    age = 2026 - year
    age_factor = max(0.72, 1.12 - age*0.012)
    premium = brand_factor[brand]
    cf = cond_factor[condition]
    noise = lambda s: np.random.normal(1, s)
    vals = base[dtype]
    gold = max(0, vals['gold']*premium*age_factor*cf*noise(0.08))
    silver = max(0, vals['silver']*premium*age_factor*cf*noise(0.08))
    copper = max(0, vals['copper']*premium*age_factor*cf*noise(0.06))
    palladium = max(0, vals['palladium']*premium*age_factor*cf*noise(0.09))
    platinum = max(0, vals['platinum']*premium*age_factor*cf*noise(0.10))
    recovery_rate = np.clip(55 + cf*30 + (year-2008)*0.4 + np.random.normal(0,4), 45, 97)
    cost = max(3, vals['cost'] + age*0.35 + (1-cf)*12 + np.random.normal(0,2))
    richness = gold*120 + silver*0.8 + copper*0.02 + palladium*180 + platinum*35
    score = np.clip(40 + recovery_rate*0.45 + richness*0.18 - cost*0.5 + np.random.normal(0,5), 1, 100)
    rows.append([item,brand,condition,recovery_rate,dtype,year,cost,gold,silver,copper,palladium,platinum,score])

cols=['Item Name','Brand Name','Device Condition','Material Recovery Rate','Device Type','Year of Manufacturing','Cost of Recovery','Gold (g)','Silver (g)','Copper (g)','Palladium (g)','Platinum (g)','Recycling Score']
df=pd.DataFrame(rows,columns=cols)
df.to_csv(r'C:\Users\HP\OneDrive\Desktop\AIML\Practice\Datasets\E-Waste\Synthetic.csv',index=False)
print(df.head())
print('\nSaved: advanced_e_waste_dataset.csv')
```
